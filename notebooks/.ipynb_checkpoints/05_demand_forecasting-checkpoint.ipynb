# Phase 5 — Demand Forecasting ML Model
**Project:** E-commerce Sales & Demand Forecasting  
**Dataset:** Online Retail II (UCI)  
**Goal:** Forecast daily revenue using Prophet and XGBoost, then compare their performance.

---
### What we build in this notebook
1. Prepare a clean daily time-series from our EDA dataset  
2. Train a **Prophet** model (handles seasonality automatically)  
3. Train an **XGBoost** model (with lag & rolling features)  
4. Evaluate both with MAE, RMSE, MAPE  
5. Visualise actual vs forecast  
6. Compare models and save results


```
# ── Cell 1: Working Directory Setup ─────────────────────────────────────────
# We launch Jupyter from the /notebooks folder, but our data lives one level up.
# os.chdir("..") moves us to the project root so all relative paths work correctly.

import os

if os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")

print("Working directory:", os.getcwd())
```

```
# ── Cell 2: Import All Required Libraries ───────────────────────────────────
import pandas as pd               # data manipulation
import numpy as np                # numerical operations
import matplotlib.pyplot as plt   # static plots (used by Prophet components)
import warnings
warnings.filterwarnings('ignore') # keep output clean

# Prophet — Facebook/Meta open-source forecasting library
# Excellent for time series with strong weekly/yearly seasonality
from prophet import Prophet

# Scikit-learn — we use it only for evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error

# XGBoost — gradient boosting, great for tabular feature-engineered data
from xgboost import XGBRegressor

# Plotly — interactive charts (same renderer as Phase 3)
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook"  # renders inline inside classic Jupyter Notebook

print("All libraries imported successfully!")
```

```
# ── Cell 3: Load the EDA-Complete Dataset ───────────────────────────────────
# We use the enriched output from Phase 3, NOT the raw file.
# parse_dates converts InvoiceDate from plain text to a proper datetime object
# so we can do .dt.month, .dt.year etc. on it.

df = pd.read_csv(
    'data/processed/online_retail_eda_complete.csv',
    parse_dates=['InvoiceDate']
)

print(f"Dataset shape  : {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"Date range     : {df['InvoiceDate'].min().date()} to {df['InvoiceDate'].max().date()}")
print(f"Columns        : {df.columns.tolist()}")
```

```
# ── Cell 4: Build a Daily Revenue Time Series ────────────────────────────────
# Our dataset has one row PER TRANSACTION (invoice line).
# Forecasting models need one row PER DAY with the total revenue for that day.
#
# Step A: Group by calendar date, sum TotalPrice.
# Step B: Fill in missing dates (days with zero sales) so the series has no gaps.
#   Why? Models like Prophet and XGBoost need evenly-spaced, continuous data.
#   A missing Sunday would confuse the weekly seasonality detection.

# Step A — aggregate
daily_revenue = (
    df.groupby(df['InvoiceDate'].dt.date)['TotalPrice']
    .sum()
    .reset_index()
)
daily_revenue.columns = ['Date', 'Revenue']
daily_revenue['Date'] = pd.to_datetime(daily_revenue['Date'])
daily_revenue = daily_revenue.sort_values('Date').reset_index(drop=True)

print(f"Days with at least one transaction: {len(daily_revenue)}")

# Step B — reindex over full calendar, fill missing days with 0
full_range = pd.date_range(
    start=daily_revenue['Date'].min(),
    end=daily_revenue['Date'].max(),
    freq='D'
)

daily_revenue = (
    daily_revenue
    .set_index('Date')
    .reindex(full_range, fill_value=0)
    .reset_index()
)
daily_revenue.columns = ['Date', 'Revenue']

print(f"Days after filling calendar gaps  : {len(daily_revenue)}")
print(f"\nRevenue stats (GBP):")
print(daily_revenue['Revenue'].describe().round(2))
```

```
# ── Cell 5: Visualise the Daily Revenue Time Series ─────────────────────────
# ALWAYS plot your time series before modelling.
# Look for: overall trend, seasonal spikes, flat/zero regions, sudden outliers.

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=daily_revenue['Date'],
    y=daily_revenue['Revenue'],
    mode='lines',
    name='Daily Revenue',
    line=dict(color='#1976D2', width=1)
))

fig.update_layout(
    title='Daily Revenue Over Time — Full Dataset',
    xaxis_title='Date',
    yaxis_title='Revenue (GBP)',
    hovermode='x unified',
    template='plotly_white',
    height=400
)

fig.write_html('reports/figures/daily_revenue_timeseries.html')
fig.show()
print("Chart saved to reports/figures/daily_revenue_timeseries.html")
```

```
# ── Cell 6: Chronological Train / Test Split ─────────────────────────────────
# IMPORTANT: In time series we NEVER split randomly.
#
# Why? Imagine randomly picking dates — the model might train on Nov data
# while testing on Oct data. That means it can 'see the future' during training,
# which is cheating and produces unrealistically good results.
#
# Correct rule: Train on EARLIER dates, test on LATER dates. Always.
#
# We hold out the last 60 days (~2 months) as the test set.

TEST_DAYS = 60

train = daily_revenue[:-TEST_DAYS].copy()
test  = daily_revenue[-TEST_DAYS:].copy()

print(f"Training: {train['Date'].min().date()} to {train['Date'].max().date()}  ({len(train)} days)")
print(f"Test    : {test['Date'].min().date()}  to {test['Date'].max().date()}  ({len(test)} days)")
```

```
# ── Cell 7: Prepare Data for Prophet ────────────────────────────────────────
# Prophet has a strict API requirement: the input DataFrame must have exactly
# two columns named:
#   'ds' — the datestamp column (date or datetime)
#   'y'  — the numeric value to forecast
#
# We simply rename our columns to match this contract.

train_prophet = train.rename(columns={'Date': 'ds', 'Revenue': 'y'})
test_prophet  = test.rename(columns={'Date': 'ds', 'Revenue': 'y'})

print("Prophet training data — first 3 rows:")
print(train_prophet.head(3).to_string(index=False))
print(f"\nTraining rows: {len(train_prophet)}")
```

```
# ── Cell 8: Train the Prophet Model ─────────────────────────────────────────
# Prophet decomposes your time series into additive components:
#   Revenue(t) = Trend(t) + Seasonality(t) + Holidays(t) + Error(t)
#
# Think of it like this analogy:
#   A store's monthly rent = base cost + holiday surcharge + random noise
#   Prophet finds each layer automatically from historical data.
#
# Key parameters:
#   yearly_seasonality  : True  — data spans 2 years, so annual cycles are real
#   weekly_seasonality  : True  — B2B business, Mon-Fri rhythm is very clear
#   daily_seasonality   : False — we have daily (not hourly) data; enabling this
#                                 would add noise with no benefit
#   changepoint_prior_scale (0.05) — how sharply the TREND can bend at change points
#       Lower = smoother, more stable trend (preferred for this dataset)
#       Higher = trend bends aggressively (risks overfitting noise)
#   seasonality_prior_scale (10)   — how strong seasonal waves can be
#   interval_width (0.95)          — 95% confidence interval on predictions

model_prophet = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
    changepoint_prior_scale=0.05,
    seasonality_prior_scale=10,
    interval_width=0.95
)

model_prophet.fit(train_prophet)    # training happens here (10-30 seconds)

print("Prophet model trained successfully!")
```

```
# ── Cell 9: Generate Prophet Forecast ───────────────────────────────────────
# make_future_dataframe creates a table of dates to predict on.
# periods=TEST_DAYS appends 60 new rows AFTER the training period ends.
# The returned dataframe covers BOTH the training period AND the test period.

future   = model_prophet.make_future_dataframe(periods=TEST_DAYS, freq='D')
forecast = model_prophet.predict(future)

# Important columns in the output:
#   ds          — the date
#   yhat        — the point forecast (our best single-number prediction)
#   yhat_lower  — lower bound of confidence interval
#   yhat_upper  — upper bound of confidence interval

print("Last 5 forecast rows (test period):")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5).to_string(index=False))
```

```
# ── Cell 10: Evaluate Prophet on the Test Set ───────────────────────────────
# We extract predictions ONLY for the 60 test dates (dates the model never saw).
# Then we compare yhat (predicted) vs Revenue (actual).

prophet_test_rows = forecast[forecast['ds'].isin(test['Date'])]
prophet_pred      = np.clip(prophet_test_rows['yhat'].values, 0, None)
# np.clip(..., 0, None) replaces any negative predictions with 0.
# Revenue cannot be negative. Prophet can output small negatives on very quiet days.

actual = test['Revenue'].values

# ── Three standard forecast accuracy metrics ─────────────────────────────
# MAE  — Mean Absolute Error
#   Average GBP error per day. Easy to explain to non-technical stakeholders.
#   "On average the model is off by GBP X per day."
#
# RMSE — Root Mean Squared Error
#   Like MAE but large errors are penalised more (squared before averaging).
#   More sensitive to big spikes/drops. Usually RMSE > MAE.
#
# MAPE — Mean Absolute Percentage Error
#   Error as a % of actual revenue. Useful for comparing across different scales.
#   We skip days where actual=0 to avoid division by zero.

mae_prophet  = mean_absolute_error(actual, prophet_pred)
rmse_prophet = np.sqrt(mean_squared_error(actual, prophet_pred))
mask         = actual != 0
mape_prophet = np.mean(np.abs((actual[mask] - prophet_pred[mask]) / actual[mask])) * 100

print("=" * 45)
print("  PROPHET  —  Test Set Performance")
print("=" * 45)
print(f"  MAE  : GBP {mae_prophet:>10,.2f}")
print(f"  RMSE : GBP {rmse_prophet:>10,.2f}")
print(f"  MAPE :     {mape_prophet:>9.2f}%")
print("=" * 45)
```

```
# ── Cell 11: Prophet Components Plot ────────────────────────────────────────
# This is one of Prophet's most valuable outputs.
# It shows exactly what the model learned about your data:
#
#   Trend    — is overall revenue growing, shrinking, or flat over the dataset?
#   Weekly   — which days of the week are typically strongest / weakest?
#   Yearly   — which months or seasons consistently drive more revenue?
#
# These components are extremely useful for business storytelling.
# In an interview you can say: "The model independently confirmed the November
# peak and the Mon-Fri B2B pattern that we first spotted during EDA."

fig_comp = model_prophet.plot_components(forecast)
plt.suptitle("Prophet Model Components — Trend, Weekly & Yearly Seasonality",
             y=1.01, fontsize=13)
plt.tight_layout()
plt.savefig('reports/figures/prophet_components.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved to reports/figures/prophet_components.png")
```

```
# ── Cell 12: Feature Engineering for XGBoost ────────────────────────────────
# XGBoost is a powerful algorithm, but it knows NOTHING about time by itself.
# It only understands rows of numbers.
# So we manually translate temporal patterns into numeric columns it can use.
#
# We create THREE types of features:
#
# 1. CALENDAR FEATURES — encode what time-period this row represents
#    e.g. month=11 → the model learns November always runs higher
#
# 2. LAG FEATURES — what was revenue X days ago?
#    e.g. lag_7 = revenue 7 days ago (same weekday last week)
#    This gives the model a 'memory' of recent performance.
#    Analogy: A shop owner checks last week's Tuesday sales before
#    deciding how much stock to order for this Tuesday.
#
# 3. ROLLING STATISTICS — recent average and volatility
#    e.g. rolling_mean_7 = average revenue over past 7 days
#    Captures momentum: if the last week was strong, tomorrow likely is too.
#
# CRITICAL — why shift(1)?
#   We shift ALL lags and rolling windows by 1 day before computing them.
#   This ensures the model NEVER sees the current day's own revenue as a feature.
#   Using today's revenue to predict today's revenue is data leakage (cheating).

def create_xgb_features(df):
    df = df.copy()

    # Calendar features
    df['dayofweek']  = df['Date'].dt.dayofweek                              # 0=Mon 6=Sun
    df['quarter']    = df['Date'].dt.quarter                                # 1-4
    df['month']      = df['Date'].dt.month                                  # 1-12
    df['year']       = df['Date'].dt.year                                   # e.g. 2010
    df['dayofyear']  = df['Date'].dt.dayofyear                              # 1-365
    df['weekofyear'] = df['Date'].dt.isocalendar().week.astype(int)         # 1-52
    df['is_weekend'] = (df['Date'].dt.dayofweek >= 5).astype(int)           # 1 if Sat/Sun

    # Lag features
    df['lag_1']  = df['Revenue'].shift(1)    # yesterday's revenue
    df['lag_7']  = df['Revenue'].shift(7)    # same day last week
    df['lag_14'] = df['Revenue'].shift(14)   # same day 2 weeks ago
    df['lag_30'] = df['Revenue'].shift(30)   # approx same day last month

    # Rolling statistics (shift(1) to avoid leaking current day)
    df['rolling_mean_7']  = df['Revenue'].shift(1).rolling(7).mean()   # 7-day avg
    df['rolling_mean_14'] = df['Revenue'].shift(1).rolling(14).mean()  # 14-day avg
    df['rolling_std_7']   = df['Revenue'].shift(1).rolling(7).std()    # 7-day volatility

    return df

# Apply to the full dataset so rolling windows are computed correctly across the split
full_featured = create_xgb_features(daily_revenue)
full_featured  = full_featured.dropna()   # first ~30 rows will have NaN from lag_30 -> drop

print(f"Feature engineering complete")
print(f"Shape : {full_featured.shape}")
print(f"Features : {[c for c in full_featured.columns if c not in ['Date','Revenue']]}")
```

```
# ── Cell 13: XGBoost — Split & Train ────────────────────────────────────────

FEATURES = [
    'dayofweek', 'quarter', 'month', 'year', 'dayofyear',
    'weekofyear', 'is_weekend',
    'lag_1', 'lag_7', 'lag_14', 'lag_30',
    'rolling_mean_7', 'rolling_mean_14', 'rolling_std_7'
]
TARGET = 'Revenue'

# Same chronological boundary as the Prophet split
split_date = test['Date'].min()

train_xgb = full_featured[full_featured['Date'] < split_date]
test_xgb  = full_featured[full_featured['Date'] >= split_date]

X_train, y_train = train_xgb[FEATURES], train_xgb[TARGET]
X_test,  y_test  = test_xgb[FEATURES],  test_xgb[TARGET]

print(f"XGBoost — Train rows: {len(X_train)}  |  Test rows: {len(X_test)}")

# ── XGBoost hyperparameters ──────────────────────────────────────────────────
# n_estimators     : number of trees to build sequentially (each corrects prior errors)
# learning_rate    : how much each new tree contributes (lower = more cautious)
# max_depth        : max depth of each tree (5 is a safe default; deeper risks overfitting)
# subsample        : fraction of training rows used per tree (0.8 adds randomness)
# colsample_bytree : fraction of features used per tree (0.8 = 80% of columns)
# random_state     : seed for reproducibility — same result every run
# n_jobs=-1        : use all available CPU cores

model_xgb = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model_xgb.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)

print("XGBoost model trained successfully!")
```

```
# ── Cell 14: Evaluate XGBoost on the Test Set ───────────────────────────────

xgb_pred = np.clip(model_xgb.predict(X_test), 0, None)

mae_xgb  = mean_absolute_error(y_test, xgb_pred)
rmse_xgb = np.sqrt(mean_squared_error(y_test, xgb_pred))
mask_xgb = y_test.values != 0
mape_xgb = np.mean(np.abs((y_test.values[mask_xgb] - xgb_pred[mask_xgb]) / y_test.values[mask_xgb])) * 100

print("=" * 45)
print("  XGBOOST  —  Test Set Performance")
print("=" * 45)
print(f"  MAE  : GBP {mae_xgb:>10,.2f}")
print(f"  RMSE : GBP {rmse_xgb:>10,.2f}")
print(f"  MAPE :     {mape_xgb:>9.2f}%")
print("=" * 45)
```

```
# ── Cell 15: Actual vs Forecast — Comparison Chart ──────────────────────────
# We overlay Actual, Prophet, and XGBoost on the same chart for the test period.
# The closer a forecast line follows the black Actual line, the better.

fig = go.Figure()

# Actual revenue — ground truth
fig.add_trace(go.Scatter(
    x=test['Date'], y=test['Revenue'],
    mode='lines', name='Actual Revenue',
    line=dict(color='black', width=2.5)
))

# Prophet point forecast
fig.add_trace(go.Scatter(
    x=test['Date'], y=prophet_pred,
    mode='lines', name='Prophet Forecast',
    line=dict(color='#1976D2', width=2, dash='dash')
))

# Prophet 95% confidence band (shaded region)
# We create a filled area by tracing upper bound forward, then lower bound backward.
fig.add_trace(go.Scatter(
    x=list(prophet_test_rows['ds']) + list(prophet_test_rows['ds'][::-1]),
    y=list(np.clip(prophet_test_rows['yhat_upper'], 0, None)) +
      list(np.clip(prophet_test_rows['yhat_lower'], 0, None)[::-1]),
    fill='toself',
    fillcolor='rgba(25, 118, 210, 0.12)',
    line=dict(color='rgba(0,0,0,0)'),
    name='Prophet 95% CI'
))

# XGBoost forecast
fig.add_trace(go.Scatter(
    x=test_xgb['Date'], y=xgb_pred,
    mode='lines', name='XGBoost Forecast',
    line=dict(color='#E53935', width=2, dash='dot')
))

fig.update_layout(
    title='Actual Revenue vs Forecasts — Test Period (Last 60 Days)',
    xaxis_title='Date',
    yaxis_title='Revenue (GBP)',
    hovermode='x unified',
    template='plotly_white',
    height=450,
    legend=dict(orientation='h', y=-0.2)
)

fig.write_html('reports/figures/model_comparison.html')
fig.show()
print("Saved to reports/figures/model_comparison.html")
```

```
# ── Cell 16: Side-by-Side Metric Comparison ─────────────────────────────────
# This single table summarises which model performed better on the test set.
# Lower MAE / RMSE / MAPE = better forecast accuracy.

comparison = pd.DataFrame({
    'Model'    : ['Prophet', 'XGBoost'],
    'MAE (GBP)': [round(mae_prophet, 2),  round(mae_xgb, 2)],
    'RMSE(GBP)': [round(rmse_prophet, 2), round(rmse_xgb, 2)],
    'MAPE (%)' : [round(mape_prophet, 2), round(mape_xgb, 2)]
})

print("\n" + "=" * 50)
print("  MODEL COMPARISON — 60-Day Test Set")
print("=" * 50)
print(comparison.to_string(index=False))
print("=" * 50)

winner = 'Prophet' if mae_prophet < mae_xgb else 'XGBoost'
print(f"\nLower MAE winner: {winner}")
```

```
# ── Cell 17: XGBoost Feature Importance ─────────────────────────────────────
# XGBoost scores each feature by how much it reduced prediction error
# across all trees. Higher importance = more useful to the model.
#
# This is great for business storytelling:
# "The model relied most on recent revenue momentum (rolling averages and
# same-weekday lags), which confirms that short-term demand is self-reinforcing
# — a strong week tends to be followed by another strong week."

feat_df = (
    pd.DataFrame({'Feature': FEATURES, 'Importance': model_xgb.feature_importances_})
    .sort_values('Importance', ascending=False)
    .reset_index(drop=True)
)

fig2 = px.bar(
    feat_df, x='Importance', y='Feature', orientation='h',
    title='XGBoost — Feature Importance (What the Model Relied on Most)',
    color='Importance',
    color_continuous_scale='Blues',
    text=feat_df['Importance'].round(3)
)
fig2.update_layout(
    template='plotly_white',
    yaxis={'categoryorder': 'total ascending'},
    height=500,
    coloraxis_showscale=False
)

fig2.write_html('reports/figures/feature_importance.html')
fig2.show()
print("Saved to reports/figures/feature_importance.html")
```

```
# ── Cell 18: Save Forecast Results to CSV ───────────────────────────────────
# Export actual revenue + both model forecasts for the test period.
# This file can later be loaded into Power BI for the dashboard (Phase 6).

results = test[['Date', 'Revenue']].copy().reset_index(drop=True)
results.rename(columns={'Revenue': 'Actual_Revenue'}, inplace=True)

# Prophet predictions (directly aligned to test dates)
results['Prophet_Forecast'] = prophet_pred

# XGBoost predictions (may have fewer rows due to lag NaN dropping)
xgb_out = test_xgb[['Date']].copy().reset_index(drop=True)
xgb_out['XGBoost_Forecast'] = xgb_pred
results  = results.merge(xgb_out, on='Date', how='left')

results.to_csv('data/processed/forecast_results.csv', index=False)

print("Forecast results saved to data/processed/forecast_results.csv")
print(f"Shape: {results.shape}")
print(results.head(5).to_string(index=False))
```

## Phase 5 Complete — What We Built

| Step | What we did |
|------|-------------|
| Daily aggregation | Collapsed transaction rows into one revenue value per day |
| Gap filling | Ensured every calendar date exists (zero-revenue days included) |
| Train/test split | Chronological 60-day hold-out — no data leakage |
| Prophet | Decomposition model — learned trend + weekly + yearly seasonality |
| XGBoost | Feature-engineered model with lag, rolling, and calendar features |
| Evaluation | MAE, RMSE, MAPE compared side-by-side on unseen test data |
| Visualisation | Forecast vs actual chart + feature importance chart saved |
| Export | forecast_results.csv saved for future dashboard use |

### Interview Talking Points

- *"I built two forecasting models — Prophet for interpretable seasonality decomposition, and XGBoost using engineered lag and rolling-window features — evaluated on a held-out 60-day test set using MAE, RMSE, and MAPE."*
- *"I used a chronological train/test split rather than random splitting, because random splitting leaks future information into training — which is invalid for any time series problem."*
- *"The Prophet components plot confirmed the November revenue peak and the Mon-Fri B2B weekly pattern independently — consistent with what we found during EDA."*
- *"XGBoost required manual feature engineering. I created lag features (yesterday, last week, last month) and rolling averages to give the model temporal memory, since gradient boosting has no built-in concept of time."*
