## 3.1 — Load Cleaned Data & Sanity Check

```python
# ── CELL 1: Imports ──────────────────────────────────────────────────────────
# We import everything we'll need for the entire EDA phase up front.
# This is good practice — avoids hunting for missing imports mid-analysis.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")   # suppress minor deprecation warnings
sns.set_theme(style="whitegrid")    # clean background for seaborn charts

print("✅ All libraries imported successfully")
```

```python
# ── CELL 2: Load Cleaned Data ────────────────────────────────────────────────
# We load the cleaned CSV that was exported at the end of Phase 2.
# parse_dates ensures InvoiceDate is read as datetime, not plain text.

df = pd.read_csv(
    "data/processed/online_retail_cleaned.csv",
    parse_dates=["InvoiceDate"]     # read this column as datetime object
)

print(f"✅ Data loaded successfully")
print(f"   Shape    : {df.shape}  (rows, columns)")
print(f"   Columns  : {list(df.columns)}")
print(f"   Date range: {df['InvoiceDate'].min()} → {df['InvoiceDate'].max()}")
```

```python
# ── CELL 3: Sanity Check ─────────────────────────────────────────────────────
# Before doing any analysis, always verify the data looks as expected.
# This catches any issues introduced during the save/load process.

print("=== First 5 rows ===")
display(df.head())

print("\n=== Data Types ===")
print(df.dtypes)

print("\n=== Null Values ===")
print(df.isnull().sum())

print("\n=== Basic Stats ===")
display(df[["Quantity", "Price", "TotalPrice"]].describe().round(2))
```

```python
# ── CELL 4: Ensure Time Feature Columns Exist ────────────────────────────────
# Phase 2 added these columns. We verify they're present.
# If they're missing for any reason, we recreate them here.

required_cols = ["Year", "Month", "Day", "DayOfWeek", "Hour", "WeekOfYear", "TotalPrice"]

missing = [col for col in required_cols if col not in df.columns]

if missing:
    print(f"⚠️ Missing columns: {missing} — recreating them now...")
    df["Year"]       = df["InvoiceDate"].dt.year
    df["Month"]      = df["InvoiceDate"].dt.month
    df["Day"]        = df["InvoiceDate"].dt.day
    df["Hour"]       = df["InvoiceDate"].dt.hour
    df["WeekOfYear"] = df["InvoiceDate"].dt.isocalendar().week.astype(int)
    df["TotalPrice"] = df["Quantity"] * df["Price"]

    # DayOfWeek as readable names, not integers
    day_map = {0:"Mon", 1:"Tue", 2:"Wed", 3:"Thu", 4:"Fri", 5:"Sat", 6:"Sun"}
    df["DayOfWeek"] = df["InvoiceDate"].dt.dayofweek.map(day_map)
    print("✅ Time feature columns recreated")
else:
    print("✅ All required columns present — no action needed")

# Also create YearMonth for grouping by month in charts
# Period type allows proper chronological sorting
df["YearMonth"] = df["InvoiceDate"].dt.to_period("M")
print("✅ YearMonth column ready")
```

## 3.2 — Sales Trends (Monthly + Day of Week)

```python
# ── CELL 5: Monthly Revenue Trend ───────────────────────────────────────────
# We group all transactions by YearMonth and sum TotalPrice.
# This gives us a time-series view of revenue over the dataset's full period.
# Key question: are there seasonal peaks? Are sales growing over time?

monthly_revenue = (
    df.groupby("YearMonth")["TotalPrice"]
    .sum()
    .reset_index()
)

# Convert Period to string for Plotly (Plotly can't plot Period type directly)
monthly_revenue["YearMonth"] = monthly_revenue["YearMonth"].astype(str)
monthly_revenue["TotalPrice"] = monthly_revenue["TotalPrice"].round(2)

fig_monthly = px.line(
    monthly_revenue,
    x="YearMonth",
    y="TotalPrice",
    title="Monthly Revenue Trend",
    labels={"YearMonth": "Month", "TotalPrice": "Total Revenue (£)"},
    markers=True
)
fig_monthly.update_layout(xaxis_tickangle=-45)
fig_monthly.write_html("reports/figures/monthly_revenue.html")
fig_monthly.show()
print("✅ Saved: reports/figures/monthly_revenue.html")
```

```python
# ── CELL 6: Revenue by Day of Week ──────────────────────────────────────────
# Which days generate the most revenue?
# In B2C (consumer) retail you'd expect weekends to be high.
# In B2B (wholesale) you'd expect weekdays — procurement happens during work hours.
# This chart tells us which type of business this is.

# Define correct weekday order (Mon → Sun)
day_order = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

dow_revenue = (
    df.groupby("DayOfWeek")["TotalPrice"]
    .sum()
    .reindex(day_order)    # force Mon-Sun order instead of alphabetical
    .reset_index()
)
dow_revenue.columns = ["DayOfWeek", "TotalRevenue"]
dow_revenue["TotalRevenue"] = dow_revenue["TotalRevenue"].round(2)

fig_dow = px.bar(
    dow_revenue,
    x="DayOfWeek",
    y="TotalRevenue",
    title="Revenue by Day of Week",
    labels={"DayOfWeek": "Day", "TotalRevenue": "Total Revenue (£)"},
    color="TotalRevenue",
    color_continuous_scale="Blues",
    text="TotalRevenue"
)
fig_dow.write_html("reports/figures/revenue_by_dow.html")
fig_dow.show()
print("✅ Saved: reports/figures/revenue_by_dow.html")
```

## 3.3 — Top Products & Countries

```python
# ── CELL 7: Filter Out Non-Product Entries ───────────────────────────────────
# 'Manual' = system-level accounting adjustment, not a real product.
# 'DOTCOM POSTAGE' / 'POSTAGE' = shipping charges, not physical goods.
# Including these would make them appear as 'top products' which is misleading.
# A good analyst always questions whether entries are real business events.

non_products = ["Manual", "DOTCOM POSTAGE", "POSTAGE", "Adjust bad debt"]

df_products_only = df[~df["Description"].isin(non_products)]
# ~ means NOT — keep rows where Description is NOT in the list

print(f"Rows before filter : {len(df):,}")
print(f"Rows after filter  : {len(df_products_only):,}")
print(f"Rows removed       : {len(df) - len(df_products_only):,}")
```

```python
# ── CELL 8: Top 10 Products by Revenue ──────────────────────────────────────
# Group by product name, sum their revenue, keep top 10.
# This reveals which SKUs (Stock Keeping Units) drive the business.
# In B2B wholesale, a small number of products = majority of revenue (Pareto principle).

top_products = (
    df_products_only.groupby("Description")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
top_products.columns = ["Product", "TotalRevenue"]
top_products["TotalRevenue"] = top_products["TotalRevenue"].round(2)

print(top_products)

fig_prod = px.bar(
    top_products,
    x="TotalRevenue",
    y="Product",
    orientation="h",
    title="Top 10 Products by Total Revenue",
    labels={"TotalRevenue": "Total Revenue (£)", "Product": ""},
    color="TotalRevenue",
    color_continuous_scale="Blues",
    text="TotalRevenue"
)
fig_prod.update_layout(yaxis={"categoryorder": "total ascending"})
fig_prod.write_html("reports/figures/top10_products.html")
fig_prod.show()
print("✅ Saved: reports/figures/top10_products.html")
```

```python
# ── CELL 9: Top 10 Countries by Revenue (Excl. UK) ───────────────────────────
# UK is the home market and accounts for ~80-85% of total revenue.
# Including it would visually crush all other countries in the chart.
# We show UK revenue separately as context, then chart international markets.

uk_revenue = df[df["Country"] == "United Kingdom"]["TotalPrice"].sum()
print(f"🇬🇧 UK Total Revenue: £{uk_revenue:,.2f}")

top_countries = (
    df[df["Country"] != "United Kingdom"]
    .groupby("Country")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
top_countries.columns = ["Country", "TotalRevenue"]
top_countries["TotalRevenue"] = top_countries["TotalRevenue"].round(2)

# Add revenue share % among non-UK countries
total_non_uk = top_countries["TotalRevenue"].sum()
top_countries["RevenueShare%"] = (top_countries["TotalRevenue"] / total_non_uk * 100).round(2)
print(top_countries)

fig_ctry = px.bar(
    top_countries,
    x="TotalRevenue",
    y="Country",
    orientation="h",
    title="Top 10 Countries by Revenue (Excluding UK)",
    labels={"TotalRevenue": "Total Revenue (£)", "Country": ""},
    color="TotalRevenue",
    color_continuous_scale="Greens",
    text="TotalRevenue"
)
fig_ctry.update_layout(yaxis={"categoryorder": "total ascending"})
fig_ctry.write_html("reports/figures/top10_countries.html")
fig_ctry.show()
print("✅ Saved: reports/figures/top10_countries.html")
```

## 3.4 — Customer Behavior & RFM Analysis

```python
# ── CELL 10: RFM Setup — Snapshot Date ──────────────────────────────────────
# RFM = Recency, Frequency, Monetary
# It scores every customer on three dimensions:
#   Recency   — how recently they bought (lower days = better)
#   Frequency — how many unique orders they placed (higher = better)
#   Monetary  — how much they spent in total (higher = better)
#
# Snapshot date = 1 day after last purchase in dataset.
# We calculate Recency as: snapshot_date - customer's last purchase date
# Adding 1 day ensures the most recent customer gets Recency=1, not 0.

snapshot_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)
print(f"Snapshot date     : {snapshot_date}")
print(f"Dataset date range: {df['InvoiceDate'].min()} → {df['InvoiceDate'].max()}")
```

```python
# ── CELL 11: Build RFM Table ─────────────────────────────────────────────────
# We aggregate at Customer ID level.
# Each row = one unique customer with their R, F, M values.

rfm = (
    df.groupby("Customer ID")
    .agg(
        Recency   = ("InvoiceDate", lambda x: (snapshot_date - x.max()).days),
        # Days since their most recent purchase

        Frequency = ("Invoice",     "nunique"),
        # Number of unique orders (not rows — one order can have many items)

        Monetary  = ("TotalPrice",  "sum")
        # Total amount spent across all their orders
    )
    .reset_index()
)

rfm["Monetary"] = rfm["Monetary"].round(2)

print(f"Total unique customers: {len(rfm):,}")
print("\nFirst 10 rows:")
display(rfm.head(10))
print("\nRFM Summary Statistics:")
display(rfm[["Recency", "Frequency", "Monetary"]].describe().round(2))
```

```python
# ── CELL 12: Score Each Customer 1–5 ────────────────────────────────────────
# We use quintiles (5 equal-sized groups) to score each R, F, M dimension.
# Score 5 = best, Score 1 = worst.
#
# Important: Recency scoring is REVERSED.
#   Low recency (bought recently) = good = score 5
#   High recency (bought long ago) = bad  = score 1
#
# rank(method='first') is used for Frequency and Monetary
# to handle ties — many customers may have the same values.

rfm["R_Score"] = pd.qcut(rfm["Recency"], q=5, labels=[5, 4, 3, 2, 1])
# Labels reversed: small recency → label 5 (best)

rfm["F_Score"] = pd.qcut(rfm["Frequency"].rank(method="first"), q=5, labels=[1, 2, 3, 4, 5])
rfm["M_Score"] = pd.qcut(rfm["Monetary"].rank(method="first"),  q=5, labels=[1, 2, 3, 4, 5])

# Combine into a 3-digit segment string e.g. '555' = best customer
rfm["RFM_Segment"] = (
    rfm["R_Score"].astype(str) +
    rfm["F_Score"].astype(str) +
    rfm["M_Score"].astype(str)
)

# Overall score = average of three
rfm["RFM_Score"] = (
    rfm["R_Score"].astype(int) +
    rfm["F_Score"].astype(int) +
    rfm["M_Score"].astype(int)
) / 3

display(rfm[["Customer ID", "Recency", "Frequency", "Monetary",
             "R_Score", "F_Score", "M_Score", "RFM_Segment", "RFM_Score"]].head(10))
```

```python
# ── CELL 13: Label Customer Segments ────────────────────────────────────────
# Based on the overall RFM_Score, assign human-readable business labels.
# These labels are what marketing teams actually use to design campaigns.

def label_segment(score):
    if score >= 4.5:   return "Champions"           # best customers
    elif score >= 4.0: return "Loyal Customers"
    elif score >= 3.0: return "Potential Loyalists"
    elif score >= 2.0: return "At Risk"              # going quiet
    else:              return "Lost"                 # disengaged

rfm["Segment"] = rfm["RFM_Score"].apply(label_segment)

segment_counts = rfm["Segment"].value_counts().reset_index()
segment_counts.columns = ["Segment", "CustomerCount"]
print(segment_counts)

fig_rfm = px.bar(
    segment_counts,
    x="Segment",
    y="CustomerCount",
    title="Customer Segments by RFM Score",
    color="Segment",
    text="CustomerCount",
    color_discrete_sequence=px.colors.qualitative.Set2
)
fig_rfm.write_html("reports/figures/rfm_segments.html")
fig_rfm.show()
print("✅ Saved: reports/figures/rfm_segments.html")
```

## 3.5 — Outlier Detection

```python
# ── CELL 14: Boxplots ────────────────────────────────────────────────────────
# A boxplot visually shows the spread of a column and highlights outliers.
# The box = middle 50% of data (Q1 to Q3)
# The line inside = median
# The whiskers = extend to 1.5 × IQR beyond Q1/Q3
# The dots beyond whiskers = outliers

cols_to_check = ["Quantity", "Price", "TotalPrice"]

fig_box, axes = plt.subplots(1, 3, figsize=(16, 5))

for i, col in enumerate(cols_to_check):
    sns.boxplot(y=df[col], ax=axes[i], color="steelblue")
    axes[i].set_title(f"Boxplot: {col}", fontsize=13)
    axes[i].set_ylabel(col)

plt.suptitle("Outlier Detection — Boxplots", fontsize=15, y=1.02)
plt.tight_layout()
plt.savefig("reports/figures/outlier_boxplots.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved: reports/figures/outlier_boxplots.png")
```

```python
# ── CELL 15: IQR Method — Count Outliers ────────────────────────────────────
# IQR (Interquartile Range) = Q3 - Q1 = the spread of the middle 50%.
# Tukey's rule: anything outside Q1 - 1.5×IQR or Q3 + 1.5×IQR = outlier.
# Think of it as: if most orders are between £10-£200,
# a £50,000 single order is an outlier — statistically extreme.

print("=== Outlier Count per Column (IQR Method) ===\n")

for col in cols_to_check:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    pct = (len(outliers) / len(df)) * 100

    print(f"📊 {col}:")
    print(f"   Q1={Q1:.2f}  |  Q3={Q3:.2f}  |  IQR={IQR:.2f}")
    print(f"   Lower fence={lower:.2f}  |  Upper fence={upper:.2f}")
    print(f"   ⚠️  Outlier rows: {len(outliers):,}  ({pct:.2f}% of data)\n")
```

```python
# ── CELL 16: Inspect Top Extreme Transactions ────────────────────────────────
# Before deciding what to do with outliers, we LOOK at them.
# If they have valid Invoice numbers and Customer IDs → they're real bulk orders.
# If they look suspicious (missing Customer ID, odd description) → data error.

print("=== Top 10 Highest TotalPrice Transactions ===\n")
top_txn = (
    df[["Invoice", "Customer ID", "Description",
        "Quantity", "Price", "TotalPrice", "Country"]]
    .sort_values("TotalPrice", ascending=False)
    .head(10)
)
display(top_txn)
```

```python
# ── CELL 17: Flag Outliers (Never Silently Delete) ───────────────────────────
# Best practice = flag, don't drop.
# Why? Because for demand forecasting, extreme orders DID happen.
# If we delete them, the model will never learn to handle peak demand.
# Flagging lets Power BI or SQL exclude them for certain analyses
# while keeping them available for others (like revenue totals).

Q1_tp  = df["TotalPrice"].quantile(0.25)
Q3_tp  = df["TotalPrice"].quantile(0.75)
IQR_tp = Q3_tp - Q1_tp
upper_fence_tp = Q3_tp + 1.5 * IQR_tp

df["IsPriceOutlier"] = df["TotalPrice"] > upper_fence_tp

total_flagged    = df["IsPriceOutlier"].sum()
pct_flagged      = df["IsPriceOutlier"].mean() * 100
outlier_revenue  = df[df["IsPriceOutlier"]]["TotalPrice"].sum()
total_revenue    = df["TotalPrice"].sum()
revenue_share    = (outlier_revenue / total_revenue) * 100

print(f"✅ 'IsPriceOutlier' column added to dataframe")
print(f"   Rows flagged    : {total_flagged:,}  ({pct_flagged:.2f}% of rows)")
print(f"   Outlier revenue : £{outlier_revenue:,.2f}  ({revenue_share:.2f}% of total revenue)")
print()
print("💡 Key Insight: Even if outlier ROWS are few, they often carry")
print("   disproportionate revenue — very typical in B2B wholesale.")
```

## 3.6 — Correlation & Feature Insights

```python
# ── CELL 18: Correlation Heatmap ─────────────────────────────────────────────
# Correlation measures how two numeric columns move together.
# Range: -1 (perfect negative) to 0 (no relation) to +1 (perfect positive)
#
# Why it matters for ML:
#   Features strongly correlated with TotalPrice = good predictors.
#   Features correlated WITH EACH OTHER = multicollinearity risk for linear models.
#   (XGBoost/Random Forest handle this better, but worth knowing.)

numeric_cols = ["Quantity", "Price", "TotalPrice",
                "Month", "DayOfWeek", "Hour", "WeekOfYear"]

# DayOfWeek is currently strings (Mon, Tue...) — convert to int for correlation
day_to_int = {"Mon": 0, "Tue": 1, "Wed": 2, "Thu": 3, "Fri": 4, "Sat": 5, "Sun": 6}
df["DayOfWeek_int"] = df["DayOfWeek"].map(day_to_int)

numeric_cols = ["Quantity", "Price", "TotalPrice",
                "Month", "DayOfWeek_int", "Hour", "WeekOfYear"]

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(
    corr_matrix,
    annot=True,           # show numbers in each cell
    fmt=".2f",            # 2 decimal places
    cmap="coolwarm",      # blue=negative, red=positive, white=zero
    center=0,
    square=True,
    linewidths=0.5
)
plt.title("Correlation Heatmap — Numeric Features", fontsize=14)
plt.tight_layout()
plt.savefig("reports/figures/correlation_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅ Saved: reports/figures/correlation_heatmap.png")
```

```python
# ── CELL 19: Revenue by Hour of Day ──────────────────────────────────────────
# Does the time of day affect order volume?
# For a B2B wholesaler, we'd expect orders during business hours (9am–5pm).
# This pattern is a useful feature for the forecasting model.

hourly = (
    df.groupby("Hour")["TotalPrice"]
    .sum()
    .reset_index()
)
hourly.columns = ["Hour", "TotalRevenue"]

fig_hour = px.line(
    hourly,
    x="Hour",
    y="TotalRevenue",
    title="Revenue by Hour of Day",
    labels={"Hour": "Hour (24h format)", "TotalRevenue": "Total Revenue (£)"},
    markers=True
)
fig_hour.write_html("reports/figures/revenue_by_hour.html")
fig_hour.show()
print("✅ Saved: reports/figures/revenue_by_hour.html")
```

```python
# ── CELL 20: Revenue by Week of Year ─────────────────────────────────────────
# Weekly granularity reveals seasonality better than monthly.
# This is especially important because Prophet (our forecasting model)
# needs to detect both weekly and yearly seasonal cycles.
# A clear dip around Week 52 (Christmas shutdown) is expected.

weekly = (
    df.groupby("WeekOfYear")["TotalPrice"]
    .sum()
    .reset_index()
)
weekly.columns = ["WeekOfYear", "TotalRevenue"]

fig_week = px.bar(
    weekly,
    x="WeekOfYear",
    y="TotalRevenue",
    title="Revenue by Week of Year",
    labels={"WeekOfYear": "Week Number", "TotalRevenue": "Total Revenue (£)"},
    color="TotalRevenue",
    color_continuous_scale="Viridis"
)
fig_week.write_html("reports/figures/revenue_by_week.html")
fig_week.show()
print("✅ Saved: reports/figures/revenue_by_week.html")
```

```python
# ── CELL 21: Average Order Value (AOV) by Country ────────────────────────────
# AOV = Total Revenue ÷ Number of Unique Orders
# This is different from total revenue.
# A country might place fewer orders but each one is massive (high-value buyer).
# AOV tells you the QUALITY of customers per market, not just quantity.
# This is a key business KPI used by e-commerce teams globally.

aov = (
    df[df["Country"] != "United Kingdom"]
    .groupby("Country")
    .agg(
        TotalRevenue = ("TotalPrice", "sum"),
        OrderCount   = ("Invoice",    "nunique")
    )
    .reset_index()
)
aov["AOV"] = (aov["TotalRevenue"] / aov["OrderCount"]).round(2)

top_aov = aov.sort_values("AOV", ascending=False).head(15)

fig_aov = px.bar(
    top_aov,
    x="Country",
    y="AOV",
    title="Average Order Value by Country (Excl. UK) — Top 15",
    labels={"AOV": "Avg Order Value (£)", "Country": ""},
    color="AOV",
    color_continuous_scale="Teal",
    text="AOV"
)
fig_aov.update_layout(xaxis_tickangle=-35)
fig_aov.write_html("reports/figures/aov_by_country.html")
fig_aov.show()
print("✅ Saved: reports/figures/aov_by_country.html")
```

```python
# ── CELL 22: Export Final EDA-Enriched Dataset ───────────────────────────────
# Save the dataframe with the new IsPriceOutlier column.
# This file becomes the input for Phase 4 (BigQuery upload).
# Always version your processed files — future you will thank present you.

output_path = "data/processed/online_retail_eda_complete.csv"
df.to_csv(output_path, index=False)

print(f"✅ Final dataset saved to: {output_path}")
print(f"   Shape  : {df.shape}")
print(f"   Columns: {list(df.columns)}")
```
